<b> Find mean, std dev, min and max of unitprice field.<b>

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
unitpr = list()
with open('Online Retail.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    row = reader.__next__()
    ind = row.index("UnitPrice")
    for row in reader:
        unitpr.append(float(row[ind]))
    print(len(unitpr))
```

```python
unitpr = np.array(unitpr)
```

```python
print ("Mean of UnitPrice is :",np.mean(unitpr))
print ("Standard Deviation of UnitPrice is :",np.std(unitpr))
print ("Minimum value of UnitPrice is :",np.min(unitpr))
print ("Maximum value of UnitPrice is :",np.max(unitpr))
```

<b> Draw histogram of unitprice field using matplotlib <b>

```python
plt.hist(unitpr) 
plt.title("Unit Price Histogram Max Frequency upto 50")
plt.xlabel('Unit Price') 
plt.ylabel('Frequency')
plt.axis([np.min(unitpr), np.max(unitpr), 0, 50]) 

plt.show()
```

<b> Find mean of unitprice group by country and plot them <b>

```python
mean_val = {}
with open('Online Retail.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    row = reader.__next__()
    indup = row.index("UnitPrice")
    indc = row.index("Country")
    for row in reader:
        if row[indc] in mean_val:
            mean_val[row[indc]].append(float(row[indup]))
        else:
            mean_val[row[indc]] = list()
            mean_val[row[indc]].append(float(row[indup]))
    print('Mean Values Country Wise\n')
    for country in mean_val.keys():
        mean_val[country] = np.mean(np.array(mean_val[country]))
        print(country.ljust(20,' '),": ", mean_val[country])
```

```python
title = "Country wise mean value of Unit Price"
bar_labels = list(mean_val.keys())
mean = list(mean_val.values())

fig = plt.figure(figsize=(12,5))
plt.xlabel('Country') 
plt.ylabel('Unit Price') 
plt.title('Mean Unit Price by Country') 

plt.bar(bar_labels, mean, width = 0.8 )
plt.xticks(bar_labels, list(mean_val.keys()), rotation=90)
plt.show()
```

<b> Find min and max of unit price group by customerID <b>

```python
cust_group = {}
with open('Online Retail.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    row = reader.__next__()
    indup = row.index("UnitPrice")
    indc = row.index("CustomerID")
    for row in reader:
        if row[indc] in cust_group:
            cust_group[row[indc]].append(float(row[indup]))
        else:
            cust_group[row[indc]] = list()
            cust_group[row[indc]].append(float(row[indup]))
    print('Minimum and Maximum Values Customer ID Wise\n')
    for cust in cust_group.keys():
        cust_group[cust] = np.array(cust_group[cust])
        print('ID ',cust," Minimum : ", np.min(cust_group[cust]),"  Maximum : ", np.max(cust_group[cust]))
```

<b> Find cumulative sales monthly and yearly and plot them <b>

```python
cumulative_sales = list()
with open('Online Retail.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    row = reader.__next__()
    ind1 = row.index("Quantity")
    ind2 = row.index("UnitPrice")
    ind3 = row.index("InvoiceDate")
    
    for row in reader:
        temp = row[ind3].split(' ')[0].split('/')
        temp = [int(i) for i in temp]
        temp.append(float(row[ind2])*float(row[ind1]))
        cumulative_sales.append(temp)
    cumulative_sales = np.array(cumulative_sales)
    print(cumulative_sales[1:5])
```

```python
years = np.unique(cumulative_sales[:,2])
```

```python
val = {}
for i in years:
    index_list = np.where(cumulative_sales[:,2] == i)
    i+=2000
    val[int(i)] = np.sum(cumulative_sales[index_list][3])
```

```python
val
```

```python
yr = val.keys()
yr_val = list(val.values())

plt.xlabel('Year') 
plt.ylabel('Sale') 
plt.title('Yearly Sales Data') 
plt.bar(yr, yr_val , width = 0.8 )
plt.show()
```

```python
month_list = ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December']
```

```python
months = np.unique(cumulative_sales[:,1])
```

```python
#for year 2010
val = {}
cumulative_sales_new = cumulative_sales[np.where(cumulative_sales[:,2] == years[0])]
for i in months:
    index_list = np.where(cumulative_sales_new[:,1] == i)
    val[int(i)] = np.sum(cumulative_sales_new[index_list,3])
```

```python
title = "Monthly value for 2010"
bar_labels = list(val.keys())
value = list(val.values())

fig = plt.figure(figsize=(12,5))
plt.xlabel('Month') 
plt.ylabel('Sales') 
plt.title(title) 

plt.bar(month_list, value, width = 0.8 )
plt.xticks(month_list, month_list, rotation=0)
plt.show()
```
